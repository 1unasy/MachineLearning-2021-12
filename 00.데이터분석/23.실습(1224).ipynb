### 2) Titanic - sns.load_dataset('titanic')
- ppt p.41

```python
# 데이터 불러오기
titanic = sns.load_dataset('titanic')
titanic.tail()
```

**a. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만드시오. category1 카테고리는 다음과 같이 정의됨**
- 1)20살이 넘으면 성별을 그대로 사용한다.
- 2)20살 미만이면 성별에 관계없이 “child”라고 한다.

```python
# 20살 미만인 승객의 성별에 child 넣기
titanic.sex[titanic.age < 20] = 'child'
```

```python
titanic.tail()
```

b. 타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값이 되도록 titanic 데이터프레임을 고치시오.

```python
titanic.age.fillna(round(titanic.age.mean(), 1), inplace=True)
```

```python
# 29.7 이 평균 -> NaN 값에 넣음
titanic.tail()
```

c. 성별, 선실(class)별, 출발지(embark_town)별 생존율을 구하시오.

```python
# 성별 생존율
titanic.groupby('sex')[['survived']].mean()
```

```python
# 선실별 생존율
titanic.groupby('class')[['survived']].mean()
```

```python
# 출발지별 생존율
titanic.groupby('embark_town')[['survived']].mean()
```

```python
# 세 부류 중 어떤 경우에 생존율이 가장 높은가?
titanic.groupby(['sex','class','embark_town'])[['survived']].mean()
```

### 3) Mile Per Gallon - sns.load_dataset('mpg')
- ppt p.42

```python
# 데이터 불러오기
mpg = sns.load_dataset('mpg')
mpg.head()
```

**a. 배기량(displacement) 대비 마력(horsepower) 열(hp_per_cc)을 추가하시오.**

```python
mpg['hp_per_cc'] = (mpg.horsepower / mpg.displacement).round(2)
```

```python
mpg.head()
```

**b. name으로부터 manufacturer(제조사)와 모델을 추출하여 새로운 열 manufacturer와 model을 추가하고, name 열은 삭제하시오.**

- **제조사 열 만들기**

```python
mpg['manufacturer'] = [n.split()[0] for n in mpg.name]
mpg.head()
```

- **모델 열 만들기**

```python
mpg['model'] = [' '.join(n.split()[1:]) for n in mpg.name]
mpg.head()
```

**c. 엔진의 실린더(cylinders) 갯수별 연비(mpg)의 평균을 구하시오.**

```python
mpg.groupby('cylinders')[['mpg']].mean().round(2)
```

**d. 생산지(origin)별 배기량 대비 마력(hp_per_cc)의 평균을 구하시오.**

```python
mpg.groupby('origin')[['hp_per_cc']].mean().round(2)
```

**e. 모델이 5개 이상인 제조사에 대하여 연비(mpg)의 평균이 가장 좋은 제조사 Top 5를 구하시오.**

```python
res = mpg.groupby('manufacturer')[['model']].count().reset_index()
model_cnt = res[res.model >= 5]
model_cnt
```
