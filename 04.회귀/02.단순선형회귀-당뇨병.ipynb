# 단순 선형회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- 당뇨병 데이터셋

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
_, axs = plt.subplots(nrows=2, ncols=5, figsize=(18,8))
for i, feature in enumerate(df.columns[:-1]):
    row, col = i//5, i%5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- BMI vs. target / BP vs. target

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
# BMI vs. target
lr.fit(df.bmi.values.reshape(-1,1), diabetes.target)
lr.score(df.bmi.values.reshape(-1,1), diabetes.target)
```

```python
# BP vs. target
lr.fit(df.bp.values.reshape(-1,1), diabetes.target)
lr.score(df.bp.values.reshape(-1,1), diabetes.target)
```

### BMI vs. target

- 훈련/테스트 데이터셋 분리(test_size=0.1)

- 회귀에서 stratify 옵션 사용 X 
    - 왜? y 값의 분포가 연속적이기 때문에 stratify 사용 불가능

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.bmi.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2021
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 훈련
lr= LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 계수, 절편, 잔차제곱의 합
lr.coef_, lr.intercept_, lr._residues
```

```python
from sklearn.metrics import r2_score        # score로 계산했던 값과 동일
pred = lr.predict(X_test)
r2_score(y_test, pred)      # lr.score(X_test, y_test) - LinearRegression에서는 이와 같이 계산
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
mse, rmse
```

- 시각화

```python
y_test[0], pred[0]
```

```python
plt.scatter(X_train, y_train)
plt.grid()
plt.xlabel('BMI')
plt.ylabel('target')
plt.title('BMI vs. Diabetes', fontsize=15)
plt.show()
```

- 회귀선 구하기

```python
xs = np.array([-0.1, 0.18])
ys = xs * lr.coef_ + lr.intercept_
ys      # 양극단의 점에 대한 y 값
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, marker='^', label='test')
# plt.scatter(X_test, pred, marker='s', label='pred')
plt.plot(xs, ys, 'r-', lw=3)
plt.grid()
plt.xlabel('BMI')
plt.ylabel('target')
plt.title('BMI vs. Diabetes', fontsize=15)
plt.legend()
plt.show()
```

### BP vs. target

```python
from sklearn.model_selection import train_test_split
X_train2, X_test2, y_train2, y_test2 = train_test_split(
    df.bp.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2021
)
X_train2.shape, X_test2.shape, y_train2.shape, y_test2.shape
```

```python
# 훈련
lr2 = LinearRegression()
lr2.fit(X_train2, y_train2)
```

```python
# 계수, 절편
# 차원이 다름
lr2.coef_, lr2.intercept_
```

```python
lr2.score(X_test, y_test)
```

- 시각화 (회귀선 포함)

```python
plt.scatter(X_train, y_train)
plt.grid()
plt.xlabel('BP')
plt.ylabel('target')
plt.title('BP vs. Diabetes', fontsize=15)
plt.show()
```

- 회귀선 구하기

```python
xs2 = np.array([-0.12, 0.14])
ys2 = xs2 * lr2.coef_ + lr2.intercept_
ys2
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X_train2, y_train2, label='train')
plt.scatter(X_test2, y_test2, marker='^', label='test')
plt.plot(xs2, ys2, 'r-', lw=3)
plt.grid()
plt.xlabel('BP')
plt.ylabel('target')
plt.title('BP vs. Diabetes', fontsize=15)
plt.legend()
plt.show()
```
