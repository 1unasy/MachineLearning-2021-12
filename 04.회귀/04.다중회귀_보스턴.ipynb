# 다중회귀 - 보스턴 주택 가격

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE'] = boston.target
df.head()
```

- 산점도와 선형회귀선

```python
sns.regplot(x='CRIM', y='PRICE', data=df)       # 음의 상관관계
plt.show()
```

- enumerate
    - 인덱스 번호와 컬렉션의 원소를 tuple형태로 반환
    <pre>
    for i, feature in enumerate(df.columns[1:-1]):                      # 12개의 컬럼
        row, col = i//4, i%4
    <pre>

```python
_, axs = plt.subplots(nrows=3, ncols=4, figsize=(16,12))            # _ == fig ; 코드에서 사용하지 않기 때문에 언더바로 표시 
for i, feature in enumerate(df.columns[1:-1]):                      # 12개의 컬럼
    row, col = i//4, i%4                                            # 12 // 4 (몫) : 3 , 12 % 4 (나머지) : 0 ??? 
    sns.regplot(x=feature, y='PRICE', data=df, ax=axs[row][col])
```

- feature별 R squared value

```python
from sklearn.linear_model import LinearRegression
```

```python
# CRIM
# skilearn 데이터 입력 [[], [], ... []] 이런 형태로 입력되어야 함
X = df.CRIM.values.reshape(-1,1)
y = boston.target
lr = LinearRegression()
lr.fit(X, y)
lr.score(X, y)
```

```python
y = boston.target
for feature in df.columns[:-1]:
    X = df[feature].values.reshape(-1,1)
    lr = LinearRegression()
    lr.fit(X,y)
    score = lr.score(X,y)
    print(f'{feature}:\t{score:.4f}')
```

- 다중선형회귀

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    boston.data, boston.target, test_size=0.1, random_state=2021
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 다중선형회귀에서는 R squared value가 무의미함 
# -> AIC BIC 사용 (R에 존재함)
# skilearn에서는 예측값을 활용할 것 
lr.score(X_test, y_test)
```

```python
# Weight
lr.coef_
```

```python
# Bias
lr.intercept_
```

- 회귀식
<br>
PRICE = (-0.11 * CRIM + 0.053 * ZN + .... - 0.55 * LSTAT) + 35.5
<br>

- 테스트 데이터

```python
X_test[0], y_test[0]
```

```python
# Weight와 Bias로 구한 예측값
np.dot(X_test[0], lr.coef_) + lr.intercept_
```

```python
# predict로 구한 예측값
lr.predict(X_test[0].reshape(1,-1))
lr.predict(X_test[:1])  # 슬라이싱 -> 2차원 데이터
```

```python
lr.coef_.shape, X_test[0].shape     
# 1차원 데이터 --> np.dot()으로 그냥 곱해줘도 된다. 
# But 여러 차원의 데이터끼리의 계산일 때는 reshape 같은 것으로 사이즈 맞춰줘야 한다.
```

```python
# 10개의 데이터에 대한 예측값과 실제값
for i in range(10):
    pred1 = np.dot(X_test[i], lr.coef_) + lr.intercept_     # y = W . x + b --> W : lr.coef_ / b : lr.intercept_ / . : np.dot() - 사이즈를 맞춰서 arg끼리 곱함
    # pred2 = lr.predict(X_test[i].reshape(1,-1))
    pred2 = lr.predict([X_test[i]])
    # pred2 = lr.predict(X_test[i:i+1])
    print(f'실제값: {y_test[i]}, 예측값1: {pred1:.2f}, 예측값2: {pred2[0]:.2f}') # 왜 pred2[0]인가? 2차원 데이터이므로 안의 값을 들고오기 위해서 indexing 해야함
```
